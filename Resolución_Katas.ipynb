```python
"""34. Crea la clase Arbol , define un árbol genérico con un tronco y ramas como atributos. Los métodos disponibles son:
crecer_tronco , nueva_rama , crecer_ramas , quitar_rama e info_arbol . El objetivo es implementar estos métodos para
manipular la estructura del árbol."""

class Árbol:

    # Metodo constructor (Se crean los atributos que tendra el arbol.)
    def __init__(self, tronco, ramas):
        self.tronco = tronco
        self.ramas = ramas

    # Metodo crecer tronco (Sumale 1 al tronco.)
    def crecer_tronco(self):
        self.tronco +=1 

    # Metodo nueva rama (Agregale un 1 a la lista, crea una rama de valor 1)
    def nueva_rama(self):
        self.ramas.append(1)

    # Metodo crecer ramas (A cada elemento de la lista sumale 1)
    def crecer_ramas(self):
        self.ramas = list(map(lambda rama: rama+1, self.ramas))

    # Metodo quitar rama (Elimina el elemento en posicion ())
    def quitar_rama(self, posición):
        self.ramas.pop(posición)

    # Metodo info arbol (Imprime la informacion)
    def info_arbol(self):
        print(f'Longitud del tronco {self.tronco}')
        print(f'Número de ramas {len(self.ramas)}') # La cantidad de elementos en la lista es la cantidad de ramas.
        print(f'Longitudes de las ramas {self.ramas}') # Cada elemento de la lista es la longitud de la rama.
```

```python
arbol = Árbol (1, [])

arbol.crecer_tronco()
arbol.nueva_rama()
arbol.crecer_ramas()
arbol.nueva_rama()
arbol.nueva_rama()
arbol.quitar_rama(2)
arbol.info_arbol()
```

```python
"""36. Crea la clase UsuarioBanco ,representa a un usuario de un banco con su nombre, saldo y si tiene o no cuenta
corriente. Proporciona métodos para realizar operaciones como retirar dinero, transferir dinero desde otro usuario y
agregar dinero al saldo."""

class UsuarioBanco:

    # Metodo constructor
    def __init__(self, nombre, saldo, cuenta_corriente):
        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente

    # Metodo retirar dinero (Se resta lo retirado y se guarda en la variable saldo. Si no hay suficiente salta el error.)
    def retirar_dinero(self, retirada):
        if self.saldo >= retirada:
           self.saldo = self.saldo - retirada
        else:
            raise ValueError ('Saldo insuficiente')
        
    # Metedo transferir dinero ( Si el saldo es suficiente se resta de su saldo y se suma al que se envia. Si no hay suficiente salta el error.)
    def transferir_dinero (self, usuario, transferencia):
        if self.saldo >= transferencia:
            self.saldo = self.saldo - transferencia
            usuario.saldo = usuario.saldo + transferencia
        else:
            raise ValueError ('Saldo insuficiente')

    # Metodo agregar dinero (Se suma la cantidad al saldo)
    def agregar_dinero (self, cantidad):
        self.saldo = self.saldo + cantidad
```

```python
"""37. Crea una función llamada procesar_texto que procesa un texto según la opción especificada: contar_palabras ,
reemplazar_palabras , eliminar_palabra . Estas opciones son otras funciones que tenemos que definir primero y llamar dentro
de la función procesar_texto ."""

def contar_palabras(texto):
    palabras_contadas = {} # Se crea el diccionario para meter la palabra y las veces que aparece.
    palabras = texto.split() # Se crea una lista con todas las palabras del texto.
    for palabra in palabras: # Para cada palabra de la lista palabras.
        if palabra in palabras_contadas: # Si la palabra esta en el diccionario. 
            palabras_contadas[palabra] +=1 #Sumale 1.
        else:                              # Si no.
            palabras_contadas[palabra] =1  # Crea la clave con valor 1.
    return palabras_contadas
def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    texto_reemplazado = texto.replace(palabra_original, palabra_nueva) # Se usa .replace para cambiar una palabra por otra.
    return texto_reemplazado

def eliminar_palabra (texto, palabra_a_eliminar):
    palabras = texto.split() # Crea una lista de las palabras del texto.
    palabras_conservadas = [] # Se crea una lista vacia para meter las palabras que no queremos eliminar.
    for palabra in palabras: # Se recorre la lista
        if palabra != palabra_a_eliminar: # Si la palabra es distinta a la que queremos eliminar.
           palabras_conservadas.append(palabra)# Metela en la lista palabras conservadas.
    texto_modificado = ' '.join(palabras_conservadas) # .join une en un string un iterable, une la lista y crea el texto nuevo.
    return texto_modificado

def procesar_texto(texto, opcion, *args): # Nos hace falta el texto, saber la opción y los argumentos o no

    if opcion == 'contar_palabras': # Si elegimos contar solo hace falta el texto.
        return contar_palabras(texto)
    
    elif opcion == 'reemplazar_palabras': # Si elegimos reemplazar hace falta el texto,
        palabra_original = args[0] # la palabra a reemplazar que es el agumento 0
        palabra_nueva = args[1]    # y la palabra que queremos poner (argumento 1).
        return reemplazar_palabras (texto, palabra_original, palabra_nueva)
    
    elif opcion == 'eliminar_palabra': # Si elegimos eliminar hace falta el texto
        palabra_a_eliminar = args[0]    # y la palabra a eliminar.
        return eliminar_palabra (texto, palabra_a_eliminar)
    
    else:
        return 'opcion no valida'


texto = 'Sabe mas el diablo por viejo que por diablo'

print(procesar_texto( texto, 'contar_palabras'))

print(procesar_texto(texto, 'reemplazar_palabras', 'diablo', 'arbol'))

print(procesar_texto(texto, 'eliminar_palabra', 'diablo'))
```

```python
# 38. Genera un programa que nos diga si es de noche, de día o tarde según la hora proporcionada por el usuario.

import datetime

hora_proporcionada = input ('Introduce la hora en formato "hh:mm:ss"') #Pedimos la hora l usuario.

h, m, s = map(int, hora_proporcionada.split(':')) 
hora = datetime.time(h, m, s) # Se convierte el texto en hora.


inicio_dia = datetime.time(7,0,0)    # Decidimos los rangos.
fin_dia = datetime.time(13,59,59)
inicio_tarde = datetime.time(14,0,0)
fin_tarde= datetime.time(20,59,59)
inicio_noche = datetime.time(21,0,0)
fin_noche = datetime.time(6,59,59)


def dia_noche(hora):
    if inicio_dia <= hora <= fin_dia: # Comparamos la hora con nuestros rangos.
        print('Es de dia')
    elif inicio_tarde <= hora <= fin_tarde:
        print('Es por la tarde')
    elif hora>= inicio_noche or hora <=fin_noche:
        print('Es de noche')
    else:
        print('No has introducido una hora correcta')

dia_noche (hora)
```

```python
"""40. Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo" , "circulo" o
"triangulo" ) y datos (una tupla con los datos necesarios para calcular el área de la figura)."""


def calculadora_area (figura, datos):
    if figura == 'rectangulo':
        base, altura = datos # Se desempaqueta la tupla datos para calcular el area.
        return base * altura
    elif figura == 'circulo':
        (radio,) = datos # Al solo necesitar un dato y tener una tupla hay que poner radio y una coma para que lo trate como tupla.
        return 3.14 * (radio **2)
    elif figura == 'triangulo':
        base, altura = datos
        return (base * altura) /2
    else:
        print('Datos erroneos')
    
area_rectangulo = calculadora_area('rectangulo', (5, 2))
print(area_rectangulo)
area_circulo = calculadora_area('circulo', (2,))
print(area_circulo)
area_triangulo = calculadora_area('triangulo', (5,2))
print(area_triangulo)
```
